```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,2,0.1)
y1 = np.sqrt(x)
y2 = x**2
y3 = np.sin(x)
plt.plot(x,y1,label = 'sqrt')
plt.plot(x,y2,label = 'square',
    c='g',ls='--',lw=2,marker='s',mfc='r')
plt.plot(x,y3,label = 'sine')
plt.legend(loc='upper left')
plt.show()
```
